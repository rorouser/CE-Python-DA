np.random.seed(42)
ventas_df = pd.DataFrame({

## 🟡 Ejercicio 5: Manejo de Datos Faltantes

Crea un DataFrame con algunos valores faltantes:

```python
datos_incompletos = pd.DataFrame({
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Precio': [10.5, np.nan, 15.0, 12.5, np.nan],
    'Stock': [50, 30, np.nan, 45, 60],
    'Categoria': ['Electrónica', 'Ropa', np.nan, 'Electrónica', 'Ropa']
})
```

Realiza las siguientes tareas:
1. Identifica cuántos valores faltantes hay en cada columna
2. Rellena los precios faltantes con la media de los precios existentes
3. Rellena el stock faltante con 0
4. Rellena la categoría faltante con 'Sin categoría'
5. Verifica que no queden valores faltantes

````python
# Crear el DataFrame
datos_incompletos = pd.DataFrame({
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Precio': [10.5, np.nan, 15.0, 12.5, np.nan],
    'Stock': [50, 30, np.nan, 45, 60],
    'Categoria': ['Electrónica', 'Ropa', np.nan, 'Electrónica', 'Ropa']
})

# Tu código aquí
print("Valores faltantes por columna:\n", datos_incompletos.isnull().sum(),'\n')
datos_incompletos['Precio'] = datos_incompletos['Precio'].fillna(datos_incompletos['Precio'].mean())
datos_incompletos['Stock'] = datos_incompletos['Stock'].fillna(0)
datos_incompletos['Categoria'] = datos_incompletos['Categoria'].fillna('Sin categoría')
print("Valores faltantes después de rellenar:\n", datos_incompletos.isnull().sum())
````

## 🟡 Ejercicio 6: Agrupación y Agregación (GroupBy)

Crea un DataFrame con datos de ventas por producto y región:

```python
ventas_region = pd.DataFrame({
    'Region': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste'],
    'Producto': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
    'Ventas': [120, 95, 135, 110, 85, 140, 100, 125],
    'Unidades': [24, 19, 27, 22, 17, 28, 20, 25]
})
```

Calcula:
1. Las ventas totales por región
2. Las ventas promedio por producto
3. El número de unidades vendidas por región
4. Crea una tabla pivote que muestre ventas por región (filas) y producto (columnas)

````python
# Crear el DataFrame
ventas_region = pd.DataFrame({
    'Region': ['Norte', 'Sur', 'Este', 'Oeste', 'Norte', 'Sur', 'Este', 'Oeste'],
    'Producto': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
    'Ventas': [120, 95, 135, 110, 85, 140, 100, 125],
    'Unidades': [24, 19, 27, 22, 17, 28, 20, 25]
})

# Tu código aquí
ventas_por_region = ventas_region.groupby('Region')['Ventas'].sum()
print("Ventas totales por región:\n", ventas_por_region,'\n')
ventas_promedio_por_producto = ventas_region.groupby('Producto')['Ventas'].sum()
print("Ventas promedio por producto:\n", ventas_promedio_por_producto,'\n')
unidades_por_region = ventas_region.groupby('Region')['Unidades'].sum()
print("Unidades vendidas por región:\n", unidades_por_region,'\n')
tabla_pivote = ventas_region.pivot_table(values='Ventas', index='Region', columns='Producto', aggfunc='sum')
print("Tabla pivote de ventas por región y producto:\n", tabla_pivote)
````

## 🔴 Ejercicio 7: Combinación de DataFrames (Merge)

Tienes dos DataFrames: uno con información de pedidos y otro con información de clientes.

```python
pedidos = pd.DataFrame({
    'PedidoID': [101, 102, 103, 104, 105],
    'ClienteID': [1, 2, 1, 3, 2],
    'Monto': [250.0, 180.5, 320.0, 95.0, 410.0],
    'Fecha': pd.to_datetime(['2025-01-15', '2025-01-16', '2025-01-16', '2025-01-17', '2025-01-18'])
})

clientes = pd.DataFrame({
    'ClienteID': [1, 2, 3, 4],
    'Nombre': ['Ana García', 'Carlos Ruiz', 'María López', 'Juan Pérez'],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']
})
```

Realiza:
1. Un merge (inner join) entre pedidos y clientes usando 'ClienteID'
2. Muestra el resultado ordenado por 'Monto' descendente
3. Calcula el monto total de pedidos por ciudad
4. Identifica el cliente que más ha gastado (total de sus pedidos)

````python
# Crear los DataFrames
pedidos = pd.DataFrame({
    'PedidoID': [101, 102, 103, 104, 105],
    'ClienteID': [1, 2, 1, 3, 2],
    'Monto': [250.0, 180.5, 320.0, 95.0, 410.0],
    'Fecha': pd.to_datetime(['2025-01-15', '2025-01-16', '2025-01-16', '2025-01-17', '2025-01-18'])
})

clientes = pd.DataFrame({
    'ClienteID': [1, 2, 3, 4],
    'Nombre': ['Ana García', 'Carlos Ruiz', 'María López', 'Juan Pérez'],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla']
})

# Tu código aquí
merged_df = pd.merge(pedidos, clientes, on='ClienteID', how='inner')
merged_df_sorted = merged_df.sort_values(by='Monto', ascending=False)
print("Pedidos y clientes combinados:\n", merged_df_sorted,'\n')
monto_por_ciudad = merged_df.groupby('Ciudad')['Monto'].sum()
print("Monto total de pedidos por ciudad:\n", monto_por_ciudad,'\n')
gasto_por_cliente = merged_df.groupby('Nombre')['Monto'].sum()
cliente_mas_gasto = gasto_por_cliente.idxmax()
print("Cliente que más ha gastado:", cliente_mas_gasto)
````

## 🔴 Ejercicio 8: Análisis de Series Temporales

Crea un DataFrame con datos de temperatura diaria durante enero 2025:

```python
np.random.seed(42)
fechas = pd.date_range('2025-01-01', '2025-01-31', freq='D')
temperaturas = 10 + 5 * np.sin(np.arange(31) * 2 * np.pi / 31) + np.random.normal(0, 2, 31)

df_temp = pd.DataFrame({
    'Fecha': fechas,
    'Temperatura': temperaturas
})
df_temp.set_index('Fecha', inplace=True)
```

Realiza:
1. Calcula la media de temperatura del mes
2. Encuentra los 5 días más cálidos
3. Calcula la temperatura promedio por semana (`.resample('W').mean()`)
4. Cuenta cuántos días la temperatura estuvo por encima de 12°C

````python
# Crear el DataFrame
np.random.seed(42)
fechas = pd.date_range('2025-01-01', '2025-01-31', freq='D')
temperaturas = 10 + 5 * np.sin(np.arange(31) * 2 * np.pi / 31) + np.random.normal(0, 2, 31)

df_temp = pd.DataFrame({
    'Fecha': fechas,
    'Temperatura': temperaturas
})
df_temp.set_index('Fecha', inplace=True)

# Tu código aquí
media_temperatura = df_temp['Temperatura'].mean()
print("Media de temperatura del mes:", media_temperatura,'\n')
dias_mas_calidos = df_temp.nlargest(5, 'Temperatura')
print("5 días más cálidos:\n", dias_mas_calidos,'\n')
temperatura_promedio_semanal = df_temp.resample('W').mean()
print("Temperatura promedio por semana:\n", temperatura_promedio_semanal,'\n')
dias_greater_12 = (df_temp['Temperatura'] > 12).sum()
print("Número de días por encima de 12°C:", dias_greater_12)
````

## 🔴 Ejercicio 9: Limpieza y Transformación de Datos

Tienes un DataFrame "sucio" con datos de productos:

```python
datos_sucios = pd.DataFrame({
    'producto': ['  laptop  ', 'MOUSE', 'Teclado', 'laptop', 'mouse', '  Monitor'],
    'precio': ['1200€', '25', '45.50€', '1150', '30€', '350'],
    'stock': ['10', '  50  ', '30', '5', 'agotado', '15'],
    'categoria': ['Informática', 'informática', 'Informática', 'Informática', 'Informática', 'pantallas']
})
```

Limpia los datos:
1. Elimina espacios en blanco de 'producto' y conviértelo a minúsculas
2. Limpia 'precio': elimina '€' y convierte a float
3. En 'stock', reemplaza 'agotado' con 0 y convierte a int
4. Estandariza 'categoria' a "Informática" o "Pantallas" (capitalizado)
5. Elimina productos duplicados (mismo nombre), manteniendo el de menor precio

````python
# Crear el DataFrame
datos_sucios = pd.DataFrame({
    'producto': ['  laptop  ', 'MOUSE', 'Teclado', 'laptop', 'mouse', '  Monitor'],
    'precio': ['1200€', '25', '45.50€', '1150', '30€', '350'],
    'stock': ['10', '  50  ', '30', '5', 'agotado', '15'],
    'categoria': ['Informática', 'informática', 'Informática', 'Informática', 'Informática', 'pantallas']
})

# Tu código aquí
datos_sucios['producto'] = datos_sucios['producto'].str.strip().str.lower()
datos_sucios['precio'] = datos_sucios['precio'].str.replace('€', '').astype(float)
datos_sucios['stock'] = datos_sucios['stock'].replace('agotado', '0').str.strip().astype(int)
datos_sucios['categoria'] = datos_sucios['categoria'].str.capitalize()
datos_limpios = datos_sucios.sort_values('precio').drop_duplicates('producto', keep='first')
print("Datos limpios:\n", datos_limpios)
````

## 🔴 Ejercicio 10: Análisis Completo de Ventas (Caso Real)

Analiza un dataset de ventas de una tienda online:

```python
np.random.seed(42)
n = 100

ventas_completo = pd.DataFrame({
    'Fecha': pd.date_range('2024-01-01', periods=n, freq='D'),
    'Producto': np.random.choice(['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares'], n),
    'Cantidad': np.random.randint(1, 10, n),
    'Precio_Unitario': np.random.choice([599.99, 19.99, 45.50, 249.99, 79.99], n),
    'Descuento': np.random.choice([0, 0.05, 0.10, 0.15], n),
    'Cliente_Tipo': np.random.choice(['Nuevo', 'Recurrente'], n, p=[0.3, 0.7])
})
```

Realiza un análisis completo:
1. Crea una columna 'Ingreso_Total' = Cantidad × Precio_Unitario × (1 - Descuento)
2. Calcula el ingreso total por tipo de producto
3. Encuentra el producto más vendido (por cantidad de unidades)
4. Calcula el ticket promedio por tipo de cliente
5. Identifica el mes con mayores ingresos
6. Calcula qué porcentaje de las ventas tuvo descuento
7. Crea un resumen ejecutivo con las métricas más importantes

````python
# Crear el DataFrame
np.random.seed(42)
n = 100

ventas_completo = pd.DataFrame({
    'Fecha': pd.date_range('2024-01-01', periods=n, freq='D'),
    'Producto': np.random.choice(['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares'], n),
    'Cantidad': np.random.randint(1, 10, n),
    'Precio_Unitario': np.random.choice([599.99, 19.99, 45.50, 249.99, 79.99], n),
    'Descuento': np.random.choice([0, 0.05, 0.10, 0.15], n),
    'Cliente_Tipo': np.random.choice(['Nuevo', 'Recurrente'], n, p=[0.3, 0.7])
})

# Tu código aquí
ventas_completo['Ingreso_Total'] = ventas_completo['Cantidad'] * ventas_completo['Precio_Unitario'] * (1 - ventas_completo['Descuento'])
ingreso_por_producto = ventas_completo.groupby('Producto')['Ingreso_Total'].sum()
print("Ingreso total por tipo de producto:\n", ingreso_por_producto,'\n')
producto_mas_vendido = ventas_completo.groupby('Producto')['Cantidad'].sum().idxmax
ticket_promedio_por_cliente = ventas_completo.groupby('Cliente_Tipo')['Ingreso_Total'].mean()
print("Ticket promedio por tipo de cliente:\n", ticket_promedio_por_cliente,'\n')
ventas_completo['Mes'] = ventas_completo['Fecha'].dt.to_period('M')
ingreso_por_mes = ventas_completo.groupby('Mes')['Ingreso_Total'].sum()
mes_mayores_ingresos = ingreso_por_mes.idxmax()
porcentaje_ventas_con_descuento = (ventas_completo[ventas_completo['Descuento'] > 0].shape[0] / ventas_completo.shape[0]) * 100
# Resumen ejecutivo
resumen_ejecutivo = {
    'Ingreso Total': ventas_completo['Ingreso_Total'].sum(),
    'Producto Más Vendido': producto_mas_vendido(),
    'Mes con Mayores Ingresos': str(mes_mayores_ingresos),
    'Porcentaje Ventas con Descuento': porcentaje_ventas_con_descuento
}
print("Resumen Ejecutivo:\n", resumen_ejecutivo)
````
