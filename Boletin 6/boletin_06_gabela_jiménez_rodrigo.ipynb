## 🟢 Ejercicio 2: Métricas de Regresión vs Clasificación (con código)

En el capítulo 3.1 vimos que regresión y clasificación usan **métricas diferentes**. Practica calculándolas.

**Parte A – Regresión** (predicción de precio de casas en miles de €):

```python
y_real_reg = np.array([200, 350, 150, 420, 300, 180, 500, 275])
y_pred_reg = np.array([210, 330, 160, 400, 310, 190, 480, 260])
```

1. Calcula **MAE** (Mean Absolute Error) con numpy: `np.mean(np.abs(...))`
2. Calcula **RMSE** (Root Mean Squared Error) con numpy: `np.sqrt(np.mean(...))`
3. Calcula **R²** con numpy usando la fórmula: $R^2 = 1 - \frac{SS_{res}}{SS_{tot}}$
4. Verifica tus resultados con sklearn (`mean_absolute_error`, `mean_squared_error`, `r2_score`)

**Parte B – Clasificación** (detección de spam: 0=legítimo, 1=spam):

```python
y_real_clf = np.array([0, 0, 1, 1, 0, 1, 0, 1, 1, 0])
y_pred_clf = np.array([0, 1, 1, 0, 0, 1, 0, 1, 0, 0])
```

5. Calcula **Accuracy**, **Precision**, **Recall** y **F1** con sklearn
6. Construye la **matriz de confusión** e interprétala
7. ¿Qué métrica priorizarías para spam y por qué?

````python
# --- Parte A: Regresión ---
y_real_reg = np.array([200, 350, 150, 420, 300, 180, 500, 275])
y_pred_reg = np.array([210, 330, 160, 400, 310, 190, 480, 260])

# Tu código aquí (calcula MAE, RMSE, R² con numpy y verifica con sklearn)
mae_np  = np.mean(np.abs(y_real_reg - y_pred_reg))
rmse_np = np.sqrt(np.mean((y_real_reg - y_pred_reg) ** 2))
ss_res  = np.sum((y_real_reg - y_pred_reg) ** 2)
ss_tot  = np.sum((y_real_reg - np.mean(y_real_reg)) ** 2)
r2_np   = 1 - ss_res / ss_tot

print("--- Parte A: Regresión ---")
print(f"MAE  (numpy):   {mae_np:.4f}")
print(f"RMSE (numpy):   {rmse_np:.4f}")
print(f"R²   (numpy):   {r2_np:.4f}")

mae_sk  = mean_absolute_error(y_real_reg, y_pred_reg)
rmse_sk = np.sqrt(mean_squared_error(y_real_reg, y_pred_reg))
r2_sk   = r2_score(y_real_reg, y_pred_reg)

print(f"\nMAE  (sklearn): {mae_sk:.4f}")
print(f"RMSE (sklearn): {rmse_sk:.4f}")
print(f"R²   (sklearn): {r2_sk:.4f}")
print("✓ Los valores coinciden\n")

# --- Parte B: Clasificación ---
y_real_clf = np.array([0, 0, 1, 1, 0, 1, 0, 1, 1, 0])
y_pred_clf = np.array([0, 1, 1, 0, 0, 1, 0, 1, 0, 0])

# Tu código aquí (calcula Accuracy, Precision, Recall, F1, Matriz de confusión)
acc  = accuracy_score(y_real_clf, y_pred_clf)
prec = precision_score(y_real_clf, y_pred_clf)
rec  = recall_score(y_real_clf, y_pred_clf)
f1   = f1_score(y_real_clf, y_pred_clf)
cm   = confusion_matrix(y_real_clf, y_pred_clf)

print("--- Parte B: Clasificación ---")
print(f"Accuracy:  {acc:.2f}")
print(f"Precision: {prec:.2f}")
print(f"Recall:    {rec:.2f}")
print(f"F1-Score:  {f1:.2f}")

print("\nMatriz de Confusión:")
print(cm)
print("\nInterpretación:")
print(f"  Verdaderos Negativos (TN): {cm[0,0]} — Legítimos correctamente clasificados")
print(f"  Falsos Positivos    (FP): {cm[0,1]} — Legítimos marcados como spam")
print(f"  Falsos Negativos    (FN): {cm[1,0]} — Spam que escapó al filtro")
print(f"  Verdaderos Positivos (TP): {cm[1,1]} — Spam correctamente detectado")

print("\n→ Para spam priorizaría Recall: es peor que un spam llegue al inbox (FN)")
print("  que un legítimo vaya a la carpeta de spam (FP).")
````

## 🟡 Ejercicio 4: Cálculo de ROI de Proyecto de ML

Se propone un proyecto para automatizar aprobación de créditos.

**Datos:**
- Inversión: 50,000€
- Volumen: 10,000 solicitudes/mes
- Objetivo: Automatizar 70% (7,000/mes)
- Costo actual (manual): 5€/solicitud
- Costo futuro (automatizado): 0.50€/solicitud
- Ingresos por solicitud: 200€

**Tareas:**
1. Ahorro mensual en costos
2. Ingresos adicionales (si mantiene tasa aprobación)
3. ROI a 6 meses
4. ¿Viable? ¿Qué otros factores considerar?

````python
# Tu código aquí
inversion       = 50_000
volumen_total   = 10_000
vol_auto        = int(volumen_total * 0.70)   # 7 000 solicitudes/mes
coste_manual    = 5.0
coste_auto      = 0.50

# 1. Ahorro mensual en costos
ahorro_mensual = vol_auto * (coste_manual - coste_auto)
print(f"1. Ahorro mensual en costos:")
print(f"   {vol_auto:,} solicitudes × ({coste_manual}€ - {coste_auto}€) = {ahorro_mensual:,.0f}€/mes\n")

# 2. Ingresos adicionales
print(f"2. Ingresos adicionales (con misma tasa de aprobación):")
print(f"   No hay ingresos adicionales: se procesan las mismas solicitudes al mismo ritmo.")
print(f"   (Se podría ganar escala si el personal liberado absorbe mayor volumen)\n")

# 3. ROI a 6 meses
beneficio_6m = ahorro_mensual * 6
roi_6m       = (beneficio_6m - inversion) / inversion * 100
payback      = inversion / ahorro_mensual

print(f"3. ROI a 6 meses:")
print(f"   Beneficio acumulado = {ahorro_mensual:,.0f}€ × 6 meses = {beneficio_6m:,.0f}€")
print(f"   ROI = ({beneficio_6m:,.0f} - {inversion:,.0f}) / {inversion:,.0f} × 100 = {roi_6m:.1f}%\n")

# 4. Viabilidad
print(f"4. ¿Viable?")
print(f"   SÍ. La inversión se recupera en {payback:.1f} meses.")
print(f"\n   Otros factores a considerar:")
print(f"   - Errores del modelo: un falso positivo aprueba un crédito que impagará.")
print(f"   - Regulación: decisiones automatizadas en crédito exigen explicabilidad (XAI).")
print(f"   - Deriva del modelo: necesitará reentrenamiento periódico (coste oculto).")
print(f"   - Riesgo reputacional si el modelo rechaza clientes por motivos injustos.")
````

## 🟡 Ejercicio 6: Desbalanceo de Clases — Del Problema a la Solución

El capítulo 3.2 explica que el desbalanceo de clases engaña con métricas como accuracy. Pero más importante: muestra **cómo solucionarlo**.

**Parte A – El engaño del modelo ingenuo:**
1. Con el dataset proporcionado, crea predicciones que siempre digan "legítima" (clase 0). Calcula accuracy, precision, recall, F1
2. ¿Por qué accuracy es engañoso aquí?

**Parte B – Tres estrategias de solución:**
3. Entrena `LogisticRegression()` sin ajustes y evalúa con F1 y recall
4. Entrena con `class_weight='balanced'` y compara
5. Aplica `SMOTE` al conjunto de train, entrena y compara

**Parte C – Comparación:**
6. ¿Qué estrategia da mejor **F1**? ¿Y mejor **recall**?
7. Si cada fraude no detectado cuesta 1000€ y cada falsa alarma cuesta 5€, ¿qué estrategia elegirías?

````python
# Dataset de fraude desbalanceado
rng = np.random.default_rng(42)
n_normal, n_fraude = 9900, 100

X_normal = rng.normal(loc=[50, 12], scale=[15, 3], size=(n_normal, 2))
X_fraude = rng.normal(loc=[80, 2], scale=[20, 1], size=(n_fraude, 2))

X = np.vstack([X_normal, X_fraude])
y = np.array([0]*n_normal + [1]*n_fraude)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Train: {len(y_train)} muestras ({(y_train==1).sum()} fraudes)")
print(f"Test:  {len(y_test)} muestras ({(y_test==1).sum()} fraudes)")

# Tu código aquí
# ── PARTE A: El engaño del modelo ingenuo ──────────────────────────────────────
y_naive = np.zeros(len(y_test), dtype=int)

print("\n--- PARTE A: Modelo Ingenuo (siempre predice 0) ---")
print(f"Accuracy:  {accuracy_score(y_test, y_naive):.4f}  ← parece muy bueno!")
print(f"Precision: {precision_score(y_test, y_naive, zero_division=0):.4f}")
print(f"Recall:    {recall_score(y_test, y_naive, zero_division=0):.4f}  ← detecta 0 fraudes")
print(f"F1-Score:  {f1_score(y_test, y_naive, zero_division=0):.4f}")
print("→ Accuracy es engañoso: el 99% de datos son clase 0; predecir siempre 0 lo maximiza.")

# ── PARTE B: Tres estrategias ──────────────────────────────────────────────────
resultados = {}
predicciones = {}

# Estrategia 1: Sin ajustes
lr_base = LogisticRegression(random_state=42, max_iter=1000)
lr_base.fit(X_train_scaled, y_train)
predicciones['Sin ajuste'] = lr_base.predict(X_test_scaled)
resultados['Sin ajuste'] = {
    'F1': f1_score(y_test, predicciones['Sin ajuste']),
    'Recall': recall_score(y_test, predicciones['Sin ajuste'])
}

# Estrategia 2: class_weight='balanced'
lr_balanced = LogisticRegression(class_weight='balanced', random_state=42, max_iter=1000)
lr_balanced.fit(X_train_scaled, y_train)
predicciones['class_weight=balanced'] = lr_balanced.predict(X_test_scaled)
resultados['class_weight=balanced'] = {
    'F1': f1_score(y_test, predicciones['class_weight=balanced']),
    'Recall': recall_score(y_test, predicciones['class_weight=balanced'])
}

# Estrategia 3: SMOTE
smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train_scaled, y_train)
lr_smote = LogisticRegression(random_state=42, max_iter=1000)
lr_smote.fit(X_train_sm, y_train_sm)
predicciones['SMOTE'] = lr_smote.predict(X_test_scaled)
resultados['SMOTE'] = {
    'F1': f1_score(y_test, predicciones['SMOTE']),
    'Recall': recall_score(y_test, predicciones['SMOTE'])
}

print("\n--- PARTE B: Comparación de Estrategias ---")
print(f"{'Estrategia':<25} {'F1':>8} {'Recall':>8}")
print("-" * 45)
for estrategia, metricas in resultados.items():
    print(f"{estrategia:<25} {metricas['F1']:>8.4f} {metricas['Recall']:>8.4f}")

# ── PARTE C: Análisis de coste ─────────────────────────────────────────────────
coste_fn = 1000
coste_fp = 5

print("\n--- PARTE C: Análisis de Coste ---")
for estrategia, y_pred in predicciones.items():
    cm_e  = confusion_matrix(y_test, y_pred)
    fn    = cm_e[1, 0]
    fp    = cm_e[0, 1]
    coste = fn * coste_fn + fp * coste_fp
    print(f"{estrategia:<25} FN={fn:3d}  FP={fp:4d}  → Coste total={coste:,.0f}€")

mejor_f1     = max(resultados, key=lambda k: resultados[k]['F1'])
mejor_recall = max(resultados, key=lambda k: resultados[k]['Recall'])
print(f"\n→ Mejor F1:     {mejor_f1}")
print(f"→ Mejor Recall: {mejor_recall}")
print(f"\n→ Con 1000€/FN vs 5€/FP, el coste de no detectar fraude es 200× mayor que")
print(f"  una falsa alarma. Elegir la estrategia con mayor Recall (SMOTE o balanced).")
````

## 🔴 Ejercicio 8: Matriz de Correlación y Selección de Features

El capítulo 3.2 explica que features muy correlacionadas entre sí son **redundantes**: aportan la misma información. Eliminarlas simplifica el modelo sin perder precisión.

**Con el dataset de viviendas proporcionado:**
1. Calcula la **matriz de correlación** y visualízala con `sns.heatmap(corr, annot=True, cmap='coolwarm')`
2. Identifica pares de features con correlación **> 0.85** entre sí
3. Para cada par redundante, ¿cuál eliminarías? (Pista: conserva la que tenga mayor correlación con `precio`)
4. Entrena un `LinearRegression` con **todas** las features y otro **sin las redundantes**
5. Compara **R²** y **tiempo** de entrenamiento. ¿Merece la pena eliminar features?

````python
# Dataset de viviendas con features redundantes
rng = np.random.default_rng(42)
n = 500

superficie       = rng.normal(100, 30, n)
habitaciones     = superficie / 25 + rng.normal(0, 0.3, n)   # Correlacionada con superficie
antiguedad       = rng.uniform(0, 50, n)
calidad          = 10 - antiguedad / 10 + rng.normal(0, 1, n) # Correlacionada con antigüedad
distancia_centro = rng.uniform(1, 30, n)

precio = (50000 + 1500 * superficie - 800 * antiguedad
          - 1200 * distancia_centro + rng.normal(0, 10000, n))

df = pd.DataFrame({
    'superficie': superficie,
    'habitaciones': habitaciones,
    'antiguedad': antiguedad,
    'calidad': calidad,
    'distancia_centro': distancia_centro,
    'precio': precio
})

print(df.describe().round(1))

# Tu código aquí
# ── 1. Matriz de correlación ───────────────────────────────────────────────────
corr = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Matriz de Correlación')
plt.tight_layout()
plt.show()

# ── 2. Pares con correlación > 0.85 entre features ────────────────────────────
features = [c for c in df.columns if c != 'precio']
corr_f   = df[features].corr()

print("\nPares con |correlación| > 0.85 entre features:")
pares_redundantes = []
for i in range(len(features)):
    for j in range(i + 1, len(features)):
        r = corr_f.iloc[i, j]
        if abs(r) > 0.85:
            print(f"  {features[i]} ↔ {features[j]}: r = {r:.3f}")
            pares_redundantes.append((features[i], features[j]))

# ── 3. Qué eliminar de cada par ────────────────────────────────────────────────
corr_precio = corr['precio'].drop('precio').abs()
print("\nCorrelación absoluta de cada feature con 'precio':")
print(corr_precio.sort_values(ascending=False).round(3))

print("\nDecisión de eliminación:")
features_eliminar = []
for f1, f2 in pares_redundantes:
    c1, c2    = corr_precio[f1], corr_precio[f2]
    eliminar  = f1 if c1 < c2 else f2
    conservar = f2 if c1 < c2 else f1
    features_eliminar.append(eliminar)
    print(f"  Par ({f1} ↔ {f2}): eliminar '{eliminar}' (r_precio={min(c1,c2):.3f}), "
          f"conservar '{conservar}' (r_precio={max(c1,c2):.3f})")

# ── 4 y 5. Entrenar y comparar ────────────────────────────────────────────────
features_reducidas = [f for f in features if f not in features_eliminar]

X_all = df[features]
X_red = df[features_reducidas]
y     = df['precio']

X_all_tr, X_all_te, y_tr, y_te = train_test_split(X_all, y, test_size=0.2, random_state=42)
X_red_tr, X_red_te, _,   _     = train_test_split(X_red, y, test_size=0.2, random_state=42)

print("\n--- Comparación de modelos ---")
for nombre, X_tr, X_te in [
    ('Todas las features', X_all_tr, X_all_te),
    ('Sin redundantes',    X_red_tr, X_red_te)
]:
    t0  = time.perf_counter()
    lr  = LinearRegression().fit(X_tr, y_tr)
    t1  = time.perf_counter()
    r2  = r2_score(y_te, lr.predict(X_te))
    print(f"  {nombre:<22}: R²={r2:.4f} | Tiempo={1000*(t1-t0):.3f} ms | Features={X_tr.shape[1]}")

print("\n→ Eliminar features redundantes mantiene prácticamente el mismo R² con un")
print("  modelo más simple, más interpretable y más rápido de entrenar.")
````

## 🔴 Ejercicio 9: Particiones Correctas de Datos

El capítulo 3.2 muestra que dividir los datos mal puede arruinar un modelo. Practica los dos casos especiales.

**Escenario A – Datos temporales (ventas diarias):**
1. Divide el dataset de ventas **aleatoriamente** con `train_test_split`. Muestra las fechas mín/máx de cada conjunto
2. ¿Aparecen fechas de test ANTES que fechas de train? (Eso es **leakage temporal**)
3. Haz una **división temporal correcta**: entrena con enero-septiembre, testea con octubre-diciembre
4. ¿Por qué la división temporal es obligatoria con datos de series temporales?

**Escenario B – Datos con grupos (múltiples registros por usuario):**
5. Divide `clientes` **aleatoriamente** y cuenta cuántos `usuario_id` aparecen en AMBOS conjuntos (usa operación de conjuntos: `set.intersection`)
6. Divide con `GroupShuffleSplit` y verifica que no hay overlap de usuarios
7. ¿Qué mide realmente cada tipo de división? (memorización vs generalización)

````python
# Escenario A: Datos temporales de ventas
rng = np.random.default_rng(42)
fechas = pd.date_range('2023-01-01', periods=365, freq='D')
ventas = pd.DataFrame({
    'fecha': fechas,
    'ventas': 100 + np.cumsum(rng.standard_normal(365))
})

# Escenario B: Datos con grupos (múltiples registros por usuario)
n_usuarios = 50
registros_por_usuario = 8
clientes = pd.DataFrame({
    'usuario_id': np.repeat(np.arange(n_usuarios), registros_por_usuario),
    'gasto': rng.uniform(10, 500, n_usuarios * registros_por_usuario),
    'frecuencia': rng.integers(1, 10, n_usuarios * registros_por_usuario),
    'abandono': np.repeat(
        rng.choice([0, 1], n_usuarios, p=[0.8, 0.2]), registros_por_usuario
    )
})

print(f"Ventas: {len(ventas)} registros diarios")
print(f"Clientes: {len(clientes)} registros, {clientes['usuario_id'].nunique()} usuarios únicos")

# Tu código aquí
# ── ESCENARIO A: Datos temporales ─────────────────────────────────────────────
# 1. División aleatoria (incorrecta)
idx_tr, idx_te = train_test_split(ventas.index, test_size=0.2, random_state=42)
train_alea = ventas.loc[idx_tr]
test_alea  = ventas.loc[idx_te]

print("\n--- A1. División ALEATORIA (incorrecta) ---")
print(f"Train: min={train_alea['fecha'].min().date()}  max={train_alea['fecha'].max().date()}")
print(f"Test:  min={test_alea['fecha'].min().date()}   max={test_alea['fecha'].max().date()}")

# 2. ¿Hay fechas de test ANTES que fechas de train?
hay_leakage = test_alea['fecha'].min() < train_alea['fecha'].max()
print(f"¿Fechas de test anteriores a fechas de train? {'SÍ → LEAKAGE TEMPORAL' if hay_leakage else 'No'}")

# 3. División temporal correcta
corte = pd.Timestamp('2023-10-01')
train_temp = ventas[ventas['fecha'] <  corte]
test_temp  = ventas[ventas['fecha'] >= corte]

print("\n--- A2. División TEMPORAL (correcta) ---")
print(f"Train: {train_temp['fecha'].min().date()} → {train_temp['fecha'].max().date()} "
      f"({len(train_temp)} registros)")
print(f"Test:  {test_temp['fecha'].min().date()}  → {test_temp['fecha'].max().date()} "
      f"({len(test_temp)} registros)")

# 4. Por qué es obligatoria
print("\n→ Con series temporales el train SIEMPRE debe preceder al test.")
print("  Si el modelo ve datos futuros durante el entreno, aprende patrones que en")
print("  producción serían imposibles de conocer. Es trampa.")

# ── ESCENARIO B: Datos con grupos ─────────────────────────────────────────────
X_c = clientes[['gasto', 'frecuencia']]
y_c = clientes['abandono']

# 5. División aleatoria → overlap de usuarios
idx_tr_c, idx_te_c = train_test_split(clientes.index, test_size=0.2, random_state=42)
usuarios_train = set(clientes.loc[idx_tr_c, 'usuario_id'])
usuarios_test  = set(clientes.loc[idx_te_c, 'usuario_id'])
overlap        = usuarios_train.intersection(usuarios_test)

print(f"\n--- B1. División ALEATORIA ---")
print(f"Usuarios en Train: {len(usuarios_train)}")
print(f"Usuarios en Test:  {len(usuarios_test)}")
print(f"Usuarios en AMBOS: {len(overlap)} → {'LEAK de usuarios (memorización)' if overlap else 'Sin overlap'}")

# 6. División con GroupShuffleSplit
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
tr_idx, te_idx = next(gss.split(clientes, y_c, groups=clientes['usuario_id']))
usuarios_train_g = set(clientes.iloc[tr_idx]['usuario_id'])
usuarios_test_g  = set(clientes.iloc[te_idx]['usuario_id'])
overlap_g        = usuarios_train_g.intersection(usuarios_test_g)

print(f"\n--- B2. División con GroupShuffleSplit ---")
print(f"Usuarios en Train: {len(usuarios_train_g)}")
print(f"Usuarios en Test:  {len(usuarios_test_g)}")
print(f"Usuarios en AMBOS: {len(overlap_g)} → {'Sin overlap ✓' if not overlap_g else 'LEAK'}")

# 7. Qué mide cada tipo
print("\n→ División aleatoria: mide si el modelo MEMORIZA el comportamiento de usuarios")
print("  ya vistos. Optimista e irreal.")
print("→ GroupShuffleSplit: mide si el modelo GENERALIZA a usuarios completamente nuevos.")
print("  Más honesto y representativo de la situación real en producción.")
````

## 📚 Resumen - Conceptos Clave Capítulos 3.1-3.2

### Cap. 3.1: Tipos y Formulación
- **Supervisado:** Regresión | Clasificación
- **No supervisado:** Clustering | Reducción
- **Refuerzo:** Recompensas
- **Formulación:** QUÉ, CUÁNDO, PARA QUÉ, COSTO

### Cap. 3.2: Datos y Sesgo
- **Sesgo selección:** Población no representada
- **Desbalanceo:** Clases desequilibradas
- **Leakage:** Información futura -> presente
- **Limpieza ≠ Calidad:** Dataset limpio puede ser terrible para ML

### Lecciones Críticas:

1. **Formulación Precisa**: Define QUÉ, CUÁNDO, PARA QUÉ, QUÉ CUESTA
2. **Sesgo de Selección**: Falta de grupos poblacionales completos
3. **Desbalanceo**: Clases con proporciones extremadamente diferentes
4. **Data Leakage**: Información futura infiltrada en features
5. **Tipos de Sesgo**: Medición, Confirmación, Supervivencia, Temporal
6. **Métricas Correctas**: En desbalanceo, Recall/F1 > Accuracy
7. **Limpieza ≠ Calidad**: Dataset sin nulls puede ser terrible para ML

### La Verdad Incómoda:
**"La mayoría de problemas en proyectos ML no están en el algoritmo. Están en los datos."**
**Los problemas de datos son invisibles pero letales: no lanzan excepciones, el modelo parece funcionar, pero falla silenciosamente en producción.**
